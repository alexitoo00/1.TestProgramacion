# Test programación
-----

## Antes de empezar

- Consigue un cronómetro de algún tipo para anotar o tempo dedicado a cada pregunta
- Asegúrate de dispoñer de tempo suficiente para realizar o test completo. Facer pausas está permitido, pero suxírese facelo seguido.
- Recoméndase dispoñer dunha folla de papel para esbozar ideas antes de programar o código final
- Si deseas utilizar algunha ferramenta extra especializada como un editor de código (VSCode, PyCharm, Tonny, etc.) ou inteligencia artificial, asegúrate de telos abertos antes de empezar.
> Sí, podes usar ChatGPT si o prefieres, pero recorda que é preferible que o que respondas o saibas defender porque o razonaches que porque cho chivou unha IA.
- Recorda que cada persona ten un proceso diferente á hora de programar e os resultados obtidos por outros pode ser igual de válido aínda que o razonasen de xeito diferente
-----

## Instruccións

- Todas as respostas deberán estar nun único arquivo. Poderase escribir o código en outro arquivo á parte para comprobar se funciona
- Activa o cronómetro antes de<!-- abrir o "toggle" e --> leer a pregunta
- Responde unha pregunta de cada vez
- Finaliza o cronómetro cando termines de escribir a resposta
- Adxunta xunto coa resposta o tempo que necesitaches e tamén se precisaches ler a "pista" (nos casos nos que a houbera)
  
    > O formato do arquivo deberá ser como segue:
    ```python
    # <Nome pregunta>
    # Tempo: 00:00 H
    # Pista: <Sí/Non>
    ................
    ... <Código> ...
    ................

    #-----
    
    # <Nome pregunta>
    # Tempo: 00:00 H
    # Pista: <Sí/Non>
    ................
    ... <Código> ...
    ................

    #-----

    ```
-----

## A ter en conta

- Tanto as preguntas como as respostas poden estar en pseudocódigo (indiferente sempre e cando se entenda o que se pretende)
- Si non recordas como se escribe a resposta en un lenguaje concreto, podes escribila con outro diferente, en pseudocódigo ou explicar o proceso
- Pode haber preguntas de desarrollo nas que non se necesite programar (Valorarase a implementación dun código ilustrativo)
- O tempo medio adicado a cada parte soe ser de 15 minutos, pero este pode variar dependendo do problema
- Espérase que as respostas dadas comproben TODOS os casos límite (excepto cando se indique o contrario)
    
    > Nota: Cando se fale de números excluiranse da definición os números imaxinarios

- As respostas poderán ser todo o elaboradas que se desexen. Por exemplo, si se pide o resultado dunha suma (5+2), o resposta pode mostrar únicamente o valor resultado (7) ou ser unha cadea de texto máis larga (5+2=7). Non se terá en conta sempre e cando o resultado sexa correcto e entendible
- Espérase que as resposta de cada exercicio sexa a máis eficiente das posibles solucións

### Ejercicio 6

Optimiza o siguiente código:

```cpp
float getCompensation(float val)
{
    bool positivo = true;
    if (val < 0)
    {
        val = val * -1.0;
        positivo = false;
    }
    if (val <= 0.4 && val >= 0.1)
    {
        val = val * comp[0];
    }

    else if (val <= 4.3 && val > 0.4)
    {
        val = val * comp[1];
    }

    else if (val <= 5 && val > 4.3)
    {
        val = val * comp[2];
    }

    else if (val <= 9.2 && val > 5)
    {
        val = val * comp[3];
    }

    else if (val <= 12.5 && val > 9.2)
    {
        val = val * comp[4];
    }

    else if (val <= 14.6 && val > 12.6)
    {
        val = val * comp[5];
    }

    if (!positivo)
        val = val * -1.0;
    return val;
}
```

> Nota: `comp[n]` é unha lista de valores para realizar a compensación dependendo do intervalo no que se encontre o valor.

Pista: [Árbol de intervalos](https://www.geeksforgeeks.org/interval-tree/)

> Existen solucións máis sencillas que un árbol de intervalos que tamén dan bos resultados
>
> Pista: Unha lista dos valores do intervalo para comprobar en cal deles se encontra
